# Defining CEM actor

For this Model-Based RL multitasking tutorial we rely on a CEM actor trained [here](https://github.com/nslyubaykin/relax_cem_example/blob/master/cem_tutorial.ipynb). Please also refer to this notebook for CEM explanation on tutorial.

We will also use pre-trained model from above described example.

# Create training and eval environments

```python
env_name = 'HalfCheetah-v2'
env = gym.make(env_name)
```

```python
env.action_space, env.action_space.high, env.action_space.low
```

```python
env.observation_space
```

```python
acs_dim = 6
obs_dim = 17
```

# Define CEM actor:

```python
actor = CEM(
    horizon=15, 
    n_candidate_sequences=1000,
    n_random_steps=1,
    n_iterations=4,
    n_elites=5,
    alpha=1,
    acs_dim=6,
    min_acs=-1,
    max_acs=1,
    device=torch.device('cpu')
)
```

# Define DeltaEnvModel

```python
ensemble_size = 5

obs_models = []
rews_models = []

for i in range(ensemble_size): 
    obs_models.append(
        ContObsContAcsToObsMLP(input_obs_dim=obs_dim, output_obs_dim=obs_dim,
                               acs_dim=acs_dim, nunits=250, nlayers=2)
    )
    rews_models.append(
        ContObsContAcsToRewsMLP(input_obs_dim=obs_dim, acs_dim=acs_dim, 
                                nunits=250, nlayers=2)
    )

model = DeltaEnvModel(
    obs_models=obs_models,
    device=torch.device('cuda'),
    learning_rate=1e-3,
    noise=0.01,
    batch_size=512,
    rews_models=rews_models,
    n_steps_per_update=500, 
    update_freq=1,
    stats_recalc_freq=500, 
    weight_decay=0.0
)
```

# Provide actor with model:

!Note: As RandomShooting actor uses dynamics model to evaluate action sequences it should have an access to it

```python
actor.set_model(model)
```

# Wrap env into Sampler:

```python
sampler = Sampler(env=env)
```

# Multi-Tasking concept:

CEM is a derivative-free optimizer which searches for optimal actions by iteratively generating random action sequences and evaluating them with fitted observation $ \hat{\Delta}_{t+1}  = f_{\theta}(s_{t}, a_{t})$ and reward $ \hat{r}_{t} = f_{\phi}(s_{t}, a_{t}) $ models. First action from elite (by total rewards sum) sequence is then executed. 

Theoretically, by replacing rewards given with model $ \hat{r}_{t} = f_{\phi}(s_{t}, a_{t}) $ by some user defined function we can alter RL agent's behavior without the need of retraining any model. 

# Visualize default behaviour specified by $ \hat{r}_{t} = f_{\phi}(s_{t}, a_{t}) $

Default reward for HalfCheetah is based on a forward horizontal velocity, therefore stimulating agent to move at a higher speed.

!Note: Sometimes actor may run 'upside down' and as such behavior is bringing the reward (up to 2k per 1k steps) it is considered 'rational' by the agent.

```python
visualize_actor(env=env, actor=actor, nsteps=200, train_sampling=False)
```

# Taking vertical position with CEM actor

Now let's design a custom reward function that would desirably make our HalfCheetah stand vertically.

To do so, pass the custom reward function to `DeltaEnvModel` through `set_reward_function()` method. Reward function should take as an input an observation np.array with a shape (BATCH_SIZE, OBS_DIM) and an action np.array with a shape (BATCH_SIZE, OBS_DIM).

The description of observation's array elements meanings may be found [here](https://www.gymlibrary.ml/environments/mujoco/half_cheetah/#observation-space)
    
To make our actor stand vertically we will design our reward function around 'body angle in radians' parameter (3rd observation element) in a following form:

$ r_{t} = (sin(rad_{t})^{2} - 1/2) * 2$

Let's plot its graph to show the intuition behind it:

```python
xs = np.arange(0, 2*np.pi, 0.01)
ys = [(np.sin(x)**2 - 0.5)*2 for x in xs]
plt.plot(xs, ys)
plt.vlines(x=[np.pi/2, 3*np.pi/2],
           ymin=-1, ymax=1,
           colors=['green', 'green'], 
           linestyles='dashed', 
           label='Vertical Position')
plt.vlines(x=[0, np.pi, 2*np.pi],
           ymin=-1, ymax=1,
           colors=['red', 'red', 'red'], 
           linestyles='dashed', 
           label='Horizontal Position')
plt.xlabel('Radians')
plt.ylabel('Reward')
plt.legend()
plt.title('Reward Plot vs Body Angle')
plt.show()
```

As we can see from the graph above the highest reward corresponds to $ \pi/2 $ and $ 3\pi/2 $ angles (vertical point). Negative reward is assigned to all horizontal regions - 0, $ \pi $ and $ 2\pi $ ensuring that agent would be forced to avoid this positions.

```python
# Define reward function

def vertical_body(obs: np.ndarray, acs: np.ndarray):
    body_angle = obs[:, 1].copy() # body angle in radians
    reward = (np.sin(body_angle)**2 - 0.5)*2
    return reward

# Pass it to the model
model.set_reward_function(reward_function=vertical_body)
```

```python
# !Note: printed reward is no longer relevant as it is not the optimised value!
visualize_actor(env=env, actor=actor, nsteps=200, train_sampling=False)
```

# Front-flip with CEM actor

Below we specify the simplest reward function to achieve front-flip with HalfCheetah. The reward here is just the sum of vertical speed and rotation speed, which aligns with a very basic intuition of performing a flip.

```python
# Define reward function

def flip(obs: np.ndarray, acs: np.ndarray):
    
    vert_speed = obs[:, 9].copy() # verical speed
    rot_speed = obs[:, 10].copy() # rotation speed
    
    reward = rot_speed + vert_speed

    return reward

# Pass it to the model
model.set_reward_function(reward_function=flip)
```

```python
# !Note: printed reward is no longer relevant as it is not the optimised value!
visualize_actor(env=env, actor=actor, nsteps=100, train_sampling=False)
```
